```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://parenting.firstcry.com/baby-names/"        # URL for firstcry to get the baby names

alphs = ['a','b','c','d','e']                           # An array containing alphabet of starting alphabet of the baby name  
for alph in alphs:
  i = 1      # Used to keep track of page number
  fileName = alph + '.txt'  #We get names starting with alphabet alph in a file
  data = open(fileName,'w')
  while True :
    page = requests.get(url + 'starting-with/' + alph + '/' + 'page/' + str(i)) #We keep changing i to go to next page
    soup = BeautifulSoup(page.content,'html.parser')
    
    if soup.find('div', class_ = 'no_record_page') != None :  # While loop stops when all pages get over 
      break
    
    for names in soup.find_all('div',class_ = 'name-info') : # Get Gender of baby
      if names.find('span','gender-icon icon left boy_icon') == None :
        gender = 'Female'
      elif names.find('span','gender-icon icon left girl_icon') == None  : 
        gender = 'Male'
      else :
        gender = 'Unisex'
        
      name = names.find('div','names').div.text  # Get name of baby
      meaning = names.find('div','names').span.text # Get meaning of baby name
      print(name + ' - ' + gender + ' - ' + meaning + '\n')  
      data.write(name + ' - ' + gender + ' - ' + meaning + '\n')
    i += 1
  data.close()
```
